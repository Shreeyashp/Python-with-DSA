```python
def tower(n, a, b, c):
  if n == 1:
    print("move 1st disk from" , a, "to" , c)
    return
  tower(n-1, a, c, b)
  print("move ", n, "th disk from", a, "to", c)
  tower(n-1, b, a, c)

```

```python
tower(1, 'source', 'auxiliary', 'destination')
```

```python
tower(2, 'source', 'auxiliary', 'destination')
```
