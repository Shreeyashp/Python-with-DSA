```python
def firstIndex(a, si, x):
  l = len(a)
  if l==0:
    return -1
  if a[si] == x:
    return si
  return firstIndex(a, si+1, x)
```

```python
a = [1,2,3,4,5]
x = 4
print(firstIndex(a, 3, x))
```

By making copy of list

```python
def firstIndex(a, x):
  l = len(a)
  if l==0:
    return -1
  
  if a[0]==x:
    return 0

  smalllist = a[1:] #making copy of list by list to check x
  smallerlistout = firstIndex(smalllist, x)

  if smallerlistout == -1:
    return -1
  else:
    return smallerlistout +1

a = [1,2,3,7,6,7,8]
firstIndex(a, 7)
```
