```python
def lastIndex(a, x):
  l = len(a)

 # base case = if list is empty return -1
  if l == 0 : #list is empty
    return -1 
  
  # Index hypo = first search on small list if x is not present here then return -1
  smalllist = a[1:]
  smallerlistoutput = lastIndex(smalllist, x)

  if smallerlistoutput != -1:
    return smallerlistoutput +1
  else:
    if a[0] == x: # check this if x is not present in small list
      return 0
    else:
      return -1

a = [1,2,3,4,5,3,6,7]
x = 3 # x means to find in list, now here to find 3 in list
lastIndex(a, x)
```

```python
def lastIndexB(a, x, si):
  l = len(a)
  if si == l:
    return -1

  smallerlistoutput = lastIndexB(a, x, si+1)
  if smallerlistoutput != -1:
    return smallerlistoutput
  else:
    if a[si] ==x:
      return si
    else:
      return -1

a = [1,4,5,4,8]
x = 4 # x means to find in list, now here to find 3 in list
lastIndexB(a, x , 0) # si=0 start index from index 0
```
