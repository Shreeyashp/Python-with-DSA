```python
from abc import ABC, abstractmethod

class automobile(ABC):
  
  def __init__(self, wheels):
    self.wheels= wheels
    print('automobile created')
  
  @abstractmethod
  def start(self):
    print('start of automobile called')
  
  @abstractmethod
  def stop(self):
    pass

  @abstractmethod
  def drive(self):
    pass

  @abstractmethod
  def get_wheels(self):
    return self.wheels

class car(automobile):

  # def __init__(self, name):
  #   print('car created')
  #   self.name = name

  def start(self):
    super().start()
    print('start of car called')

  def stop(self):
    pass

  def drive(self):
    pass

  def get_wheels(self):
    return super().get_wheels()

class bus(automobile):
  # def __init__(self, name= 'delhi bus'):
  #   print("bus created")
  #   self.name = name

  def start(self):
    pass

  def stop(self):
    pass

  def drive(self):
    pass

  def get_wheels(self):
    return super().get_wheels()
```
