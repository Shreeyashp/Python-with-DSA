Example = 
* Let Base_class1/ Parent_class be mother and Base_class2/ Parent_class be Father
* Derived/ child class be child 
* child will inherit property from both mother as well as father 


```python
class mother:
  def print(self):
    print('Print of mother called: ')

class father:
  def print(self):
    print('print of father called: ')

class child(mother, father):
  def __init__(self, name):
    self.name = name

  def printchild(self):
    print('name of child is: ', self.name)
```

```python
c=child('shree')
c.printchild()
```

```python
c=child('shreeyash')
c.print()
```

```python

class Mother:
    def __init__(self):
        self.name = "Manju"
        super().__init__()

    def print(self):
        print("Print Of Mother called")

class Father:
    def __init__(self):
        self.name = "Ajay"
                
    def print(self):
        print("Print Of Father called")

class Child(Mother,Father):
    def __init__(self):
        super().__init__()
    def print(self):
        print("Name of child is", self.name)

c = Child()
c.print()
print(Child.mro())
  
```
